## Analyse de la distribution des variables numériques

### Subtask:
Visualiser la distribution de chaque variable numérique à l'aide de boîtes à moustaches pour détecter les asymétries, les modes multiples et les éventuelles valeurs aberrantes. Nous inclurons des légendes claires pour chaque visualisation.


**Reasoning**:
To visualize the distribution of numerical variables, I will import `matplotlib.pyplot` and `seaborn`, identify the numerical columns in `df_excel` (excluding 'ID'), and then generate a box plot for each numerical column with an appropriate title.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Identifier les colonnes numériques (exclure 'ID')
numerical_cols = df_excel.select_dtypes(include=['number']).columns.tolist()
if 'ID' in numerical_cols:
    numerical_cols.remove('ID')

# Visualiser la distribution de chaque variable numérique avec des histogrammes
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.histplot(df_excel[col], kde=True)
    plt.title(f'Distribution de la colonne: {col}')
    plt.xlabel(col)
    plt.ylabel('Fréquence')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()
```

## Analyse de corrélation

### Subtask:
Calculer la matrice de corrélation entre les variables numériques pour identifier les relations linéaires. Une carte thermique (heatmap) sera générée pour visualiser ces corrélations, avec des légendes pour comprendre l'intensité et la direction des relations.


**Reasoning**:
To calculate and visualize the correlation matrix, I will use the `.corr()` method on `df_excel` to get the correlations between numerical variables, and then generate a heatmap using `seaborn.heatmap()` with annotations and a 'coolwarm' colormap for better visualization. A title will be added for clarity.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculer la matrice de corrélation
correlation_matrix = df_excel[numerical_cols].corr()

# Créer la carte thermique
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Matrice de Corrélation des Variables Numériques', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Identification des valeurs aberrantes

### Subtask:
Examiner visuellement les boîtes à moustaches ou utiliser des techniques statistiques pour identifier les valeurs aberrantes potentielles dans les variables clés, et discuter de leur impact possible sur l'analyse.


**Reasoning**:
To systematically identify outliers, I will calculate the IQR for each numerical column, define the upper and lower bounds, and then print any values falling outside these bounds. This will provide a quantitative assessment of outliers.



```python
print("Identification des valeurs aberrantes (Outliers) par la méthode IQR:")
outliers_data = {}

for col in numerical_cols:
    Q1 = df_excel[col].quantile(0.25)
    Q3 = df_excel[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df_excel[(df_excel[col] < lower_bound) | (df_excel[col] > upper_bound)][col]

    if not outliers.empty:
        outliers_data[col] = outliers
        print(f"\nOutliers dans la colonne '{col}':")
        print(outliers)
    # else:
    #     print(f"\nAucun outlier détecté dans la colonne '{col}' par la méthode IQR.")
```
