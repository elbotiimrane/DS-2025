```python
# 8. Feature Importance Analysis
def plot_feature_importance(model, feature_names, title):
    importances = model.feature_importances_

    print(f"DEBUG: Number of features in model importances: {len(importances)}")
    print(f"DEBUG: Number of feature names provided: {len(feature_names)}")

    # Determine how many features to plot (min of available importances, feature names, and top 20)
    num_features_to_plot = min(len(importances), len(feature_names), 20)

    # Get indices of top features based on importances
    indices = np.argsort(importances)[::-1][:num_features_to_plot]

    plt.figure(figsize=(12, 6))
    plt.title(title)
    plt.bar(range(num_features_to_plot),
            importances[indices],
            align="center")
    plt.xticks(range(num_features_to_plot),
               [feature_names[i] for i in indices],
               rotation=90)
    plt.tight_layout()
    plt.show()

# After training Random Forest
plot_feature_importance(best_random_forest_model, X_train.columns, "Random Forest Feature Importance")

# 9. Time Series Visualization
plt.figure(figsize=(15, 8))
plt.plot(df['Date'], df['Close'], label='Actual Price', alpha=0.7)
plt.plot(df['Date'], df['Close_30_day_MA'], label='30-day MA', alpha=0.7)
plt.fill_between(df['Date'],
                 df['Bollinger_Lower'],
                 df['Bollinger_Upper'],
                 alpha=0.2,
                 label='Bollinger Bands')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.title('Bitcoin Price with Technical Indicators')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# 10. Model comparison visualization
models = ['Linear', 'Random Forest', 'Gradient Boosting']
rmse_scores = [rmse_linear, rmse_rf, rmse_gb]
r2_scores = [r2_linear, r2_rf, r2_gb]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

# RMSE Comparison
ax1.bar(models, rmse_scores, color=['blue', 'green', 'orange'])
ax1.set_title('RMSE Comparison')
ax1.set_ylabel('RMSE')
ax1.set_xlabel('Model')

# R Comparison
ax2.bar(models, r2_scores, color=['blue', 'green', 'orange'])
ax2.set_title('R Score Comparison')
ax2.set_ylabel('R Score')
ax2.set_xlabel('Model')
ax2.set_ylim([0, 1])

plt.tight_layout()
plt.show()

# 11. Residual Analysis
def plot_residuals(y_true, y_pred, model_name):
    residuals = y_true - y_pred

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    # Residuals vs Predicted
    ax1.scatter(y_pred, residuals, alpha=0.5)
    ax1.axhline(y=0, color='r', linestyle='--')
    ax1.set_xlabel('Predicted Values')
    ax1.set_ylabel('Residuals')
    ax1.set_title(f'{model_name}: Residuals vs Predicted')

    # Residuals Distribution
    ax2.hist(residuals, bins=30, edgecolor='black')
    ax2.set_xlabel('Residuals')
    ax2.set_ylabel('Frequency')
    ax2.set_title(f'{model_name}: Residuals Distribution')

    plt.tight_layout()
    plt.show()

    # Print residual statistics
    print(f"\n{model_name} Residual Statistics:")
    print(f"Mean: {residuals.mean():.4f}")
    print(f"Std: {residuals.std():.4f}")
    print(f"Skewness: {residuals.skew():.4f}")

# Plot residuals for each model
plot_residuals(y_test, y_pred_linear, "Linear Regression")
plot_residuals(y_test, y_pred_rf, "Random Forest")
plot_residuals(y_test, y_pred_gb, "Gradient Boosting")
```
